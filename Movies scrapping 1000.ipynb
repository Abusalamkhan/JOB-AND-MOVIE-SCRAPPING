# Movies Scrapping

```python
list_movies=[]
def scrapmovies():
    c=0
    for i in range(1,51):
        url='https://www.themoviedb.org'
        urls=url+'/movie?page='+str(i)
        headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
        html_text=requests.get(urls,headers=headers).text
        soup=BeautifulSoup(html_text,'lxml')
        movies=soup.find_all('div',class_='card style_1')
        for movie in movies:
            c=c+1
            print(c,end=' ')
            name=movie.find('a')['title']
            rating=movie.find('div', class_='percent').span['class'][-1].replace('icon-r','')
            u=movie.find('h2').a['href']
            m_url="https://www.themoviedb.org"+u
            headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
            rel=requests.get(m_url,headers=headers)
            movie_html = rel.text
            movie_soup = BeautifulSoup(movie_html, 'lxml')
            m= movie_soup.find_all('div',class_='header_poster_wrapper false')
            for movie_1 in m:
                genre=movie_1.find('span',class_='genres').text.strip().replace('\xa0','')
                release_date=movie_1.find('span',class_='release').text.strip().split()[0]
                try:
                    runtime=movie_1.find('span', class_='runtime').text.strip()
                except:
                    runtime=0
                for i in movie_1.find_all('li', class_='profile'):
                    if 'Director' in i.text:
                        director = i.p.text
            
                movies_dict={
                "Name" : name,
                "Rating" : rating,
                "Genre" : genre,
                "Release Date" : release_date,
                "Runtime" : runtime,
                "Director" : director,
                "Url" : m_url
                }
                list_movies.append(movies_dict)
```

```python
scrapmovies()
```
